```python
# IMPORTAR bibliotecas
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
archivo = pd.read_csv('PROCESADO.csv', encoding='latin-1')


km = archivo["km_impulsado"]
precio = archivo["precio_de_venta"]

archivo
```

```python
# DATOS (tamaño en m^2, precio en USD)
data1 = {
    'km_': km,
    'Precio_': precio
}
```

```python
# Convertir los datos en un DataFrame
df = pd.DataFrame(data1)

# variables independientes (Año) y dependiente (precio)
X = df[['km_']]
Y = df['Precio_']

# crear un modelo de regresion lineal
modelo = LinearRegression()

# Ajustar el modelo a los datos
modelo.fit(X,Y)

# Coeficientes de la regresion
beta_0 = modelo.intercept_ # INTERCEPTO
beta_1 = modelo.coef_ # Pendientes (tamaño y habitaciones)

# Mostrar los coeficientes
print(f"Intercepto (β0): {beta_0}")
print(f"Coeficiente (km_, β1): {beta_1}")

# Visualizar la regreesion (opcional para datos simples)
plt.scatter(df['km_'], Y, color = 'blue', label = 'Datos Reales')
plt.plot(df['km_'], modelo.predict(X), color = 'red', label = 'Linea de Regresion')
plt.title('Regresion Lineal: km VS precio')
plt.xlabel('Km')
plt.ylabel('Precio (USD)')
plt.legend()
plt.show()
```

```python
test = pd.read_csv('TEST.csv', encoding='latin-1')

t_km = test["km_impulsado"]

```

```python
# Predecir precios para dos nuevos valores
#nuevo_X = np.array([t_km])  # Nuevos valores (Tamaño, Habitaciones)
nuevo_X = np.array(t_km).reshape(-1, 1)  # Convertir 't_km' a una columna con múltiples filas


predicciones = modelo.predict(nuevo_X)

# Mostrar las predicciones
print("\nPredicciones_____:")
for i, pred in enumerate(predicciones):
    print(f"Predicción {i+1} para (Tamaño: {nuevo_X[i][0]} m²): {pred:.2f} USD")
```
